# Making Individual SpecPhot Function and Testing with Vega

```python
# Vega for reference

###################### loading in wavelength and flux values
vega_wave, vega_flux = np.loadtxt('./UVOTspecphot/vega.dat',dtype=float,usecols=(0,1),unpack=True)

##################### loading in filter
vega_Filter = './UVOTspecphot/V_UVOT.txt' # for test
# input vega_wave and vega_flux into w_f_in to test. input a filter txt file as the Filter argument
# Calculates zeropoints for filter used
# inputting vega should give you a zero for magnitude in the filter used

##################### making the spectrophotometric conversion function
def vegaspecphot(wavez,fluxz,Filter):
    ##################### defining constants
    h = 6.6260755e-27
    c = 2.99792458e18
    hc = h*c # units of erg*A
    ##################### loading in filter values
    filter_lambda,filter_area = np.loadtxt(Filter,comments='#',usecols=(0,1), unpack=True)
    nonzero = np.where(filter_area > 0.0)
    filter_lambda = filter_lambda[nonzero]
    filter_area = filter_area[nonzero]
    ##################### calculating vega zeropoint for every filter from vega spectrum
    in_lambda_range = np.where((vega_wave>=min(filter_lambda))&(vega_wave<=max(filter_lambda)))
    interpolated_flux = np.interp(filter_lambda,vega_wave[in_lambda_range[0]],vega_flux[in_lambda_range[0]])
    zeropoint = round(-2.5*np.log10(np.trapz(filter_area*interpolated_flux*filter_lambda/hc,filter_lambda)),3)
    ##################### calculating magnitudes
    sp_ea = np.interp(wavez,filter_lambda,filter_area) # spectrum effective area         
    counts = np.trapz(sp_ea*fluxz*wavez/hc,wavez) # integrating under the curve using numpy
    if counts > 0:           
        vegamag = -2.5*np.log10(counts) - zeropoint # calculated magnitudes
        return vegamag
    else:
        return float("nan")

mag = vegaspecphot(vega_wave,vega_flux,vega_Filter)

print(mag)
```

# Making General SpecPhot Function

```python
############### inputs are the observational data file and the filter you want to use
def specphot(obsdata,philter):
    # load in data
    wave_sp, flux_sp = np.loadtxt(obsdata,dtype=float,usecols=(0,1),unpack=True)
    # run it through the specphot function
    vsp = vegaspecphot(wave_sp,flux_sp,philter)
    return vsp
```

```python
############### inputs are the observational data file and the filter you want to use
def kspecphot(obsdata,philter,z):
    # load in data
    wave_sp, flux_sp = np.loadtxt(obsdata,dtype=float,usecols=(0,1),unpack=True)
    kwave =  wave_sp(1+z)
    # run it through the specphot function
    vsp = vegaspecphot(kwave,flux_sp,philter)
    return vsp
```
